```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
leaderboard_df = pd.read_csv('leaderboard.csv')
leaderboard_df.head()
```

```python
leaderboard_df = leaderboard_df[:2000]
```

```python
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(10, 6))
sns.distplot(leaderboard_df["Score"], bins=10, kde=True, )
plt.xlim(0.3, 0.7)
plt.title("Distribution of Scores on the Leaderboard")
plt.axvline(x=0.4977, color='red', linestyle='--', linewidth=2)
plt.text(0.4977, 5, 'Team 4 Score', rotation=90, color='red', fontsize=12, ha='right')
```
